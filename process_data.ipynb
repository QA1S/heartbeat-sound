**Data Preparation 1: Combining set a and set b**

```python
import os
import shutil
import pandas as pd

# Define the input directories
INPUT_DIR = "/root/.cache/kagglehub/datasets/kinguistics/heartbeat-sounds/versions/1"
SET_A_DIR = os.path.join(INPUT_DIR, "set_a")
SET_B_DIR = os.path.join(INPUT_DIR, "set_b")
SET_A_METADATA = os.path.join(INPUT_DIR, "set_a.csv")
SET_B_METADATA = os.path.join(INPUT_DIR, "set_b.csv")
OUTPUT_DIR = os.path.join(INPUT_DIR, "combined_data")

# Ensure the combined folder exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Initialize an empty list to store file metadata
file_metadata = []

# Function to copy files and store metadata
def copy_files_and_collect_metadata(source_dir, dataset_label):
    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(OUTPUT_DIR, filename)

        # Copy the file to the combined folder
        if os.path.isfile(source_path):  # Ensure it's a file
            shutil.copy(source_path, dest_path)
            # Append metadata to the list
            file_metadata.append({
                "fname": os.path.basename(filename),  # Use only the file name
                "dataset": dataset_label,
                "file_path": dest_path
            })

# Copy files from set_a and set_b
copy_files_and_collect_metadata(SET_A_DIR, "a")
copy_files_and_collect_metadata(SET_B_DIR, "b")

# Load metadata files
set_a_metadata = pd.read_csv(SET_A_METADATA)
set_b_metadata = pd.read_csv(SET_B_METADATA)

# Add a column to indicate the dataset source
set_a_metadata['dataset'] = 'a'
set_b_metadata['dataset'] = 'b'

# Remove 'set_a/' or 'set_b/' from the fname column in the metadata
set_a_metadata['fname'] = set_a_metadata['fname'].apply(os.path.basename)
set_b_metadata['fname'] = set_b_metadata['fname'].apply(os.path.basename)

# Combine metadata into one DataFrame
combined_metadata = pd.concat([set_a_metadata, set_b_metadata], ignore_index=True)

# Add the `file_path` column using the copied file paths
combined_metadata['file_path'] = combined_metadata['fname'].apply(
    lambda fname: os.path.join(OUTPUT_DIR, fname)
)

# Save the combined metadata as a CSV file
combined_metadata_path = os.path.join(OUTPUT_DIR, "combined_dataset.csv")
combined_metadata.to_csv(combined_metadata_path, index=False)

print(f"Metadata saved to {combined_metadata_path}")

# Sanity check: Preview the combined CSV
if os.path.exists(combined_metadata_path):
    combined_data = pd.read_csv(combined_metadata_path)
    print("\nColumn names in combined_dataset.csv:")
    print(list(combined_data.columns))
    print("\nPreview of combined_dataset.csv:")
    print(combined_data.head())

# Sanity check: List all files in the combined_data folder
# if os.path.exists(OUTPUT_DIR):
#     print("\nFiles in combined_data folder:")
#     combined_files = os.listdir(OUTPUT_DIR)
#     for file in combined_files:
#         print(file)


```

**Data Preparation 1: Deleting all files without any label**

```python
import os
import pandas as pd

# Define paths
COMBINED_FOLDER = "/root/.cache/kagglehub/datasets/kinguistics/heartbeat-sounds/versions/1/combined_data"
METADATA_FILE = os.path.join(COMBINED_FOLDER, "combined_dataset.csv")

# Load metadata
if os.path.exists(METADATA_FILE):
    metadata = pd.read_csv(METADATA_FILE)
else:
    print(f"Metadata file {METADATA_FILE} does not exist.")
    metadata = pd.DataFrame()

# Initialize counter for deleted files
deleted_files_count = 0

# Loop through the metadata and delete files with names starting with "__" or "Bunlabelledtest"
for index, row in metadata.iterrows():
    fname = row['fname']

    # Hardcoded fixes for file paths
    if fname.startswith("Bunlabelledtest"):
        # Add another underscore to the first one
        fname = fname.replace("Bunlabelledtest_", "Bunlabelledtest__", 1)
    elif fname.startswith("__"):
        # Add "Aunlabelledtest" before the name
        fname = "Aunlabelledtest" + fname

    # Construct the modified file path
    file_path = os.path.join(COMBINED_FOLDER, fname)

    # Check if the filename starts with "__" or "Bunlabelledtest"
    if fname.startswith("Aunlabelledtest") or fname.startswith("Bunlabelledtest__"):
        # print(f"Deleting file: {file_path}")
        # Delete the file
        if os.path.exists(file_path):
            os.remove(file_path)
            deleted_files_count += 1
            # print(f"Deleted file: {file_path}")

        # Remove the row from metadata
        metadata.drop(index, inplace=True)

# Save the updated metadata back to the CSV
metadata.to_csv(METADATA_FILE, index=False)

# Print the number of remaining files in the combined_data folder
remaining_files = len([f for f in os.listdir(COMBINED_FOLDER) if f.endswith(".wav")])
metadata_rows = len(metadata)

print(f"\nDeleted {deleted_files_count} files with invalid names ('__' or 'Bunlabelledtest').")
print(f"Number of files remaining in combined_data folder: {remaining_files}")
print(f"Number of rows remaining in combined_dataset.csv: {metadata_rows}")
```

**Data Preparation 2: Formatting File Names to Match Audio Files**


```python
import os
import pandas as pd

# Define paths
COMBINED_FOLDER = "/root/.cache/kagglehub/datasets/kinguistics/heartbeat-sounds/versions/1/combined_data"
METADATA_FILE = os.path.join(COMBINED_FOLDER, "combined_dataset.csv")

# Load the existing metadata
if os.path.exists(METADATA_FILE):
    metadata = pd.read_csv(METADATA_FILE)
else:
    raise FileNotFoundError(f"Metadata file {METADATA_FILE} does not exist.")

# Function to adjust the filenames
def adjust_filename(fname):
    if fname.startswith("Btraining_"):
        # Remove the first "Btraining_" prefix
        fname = fname.replace("Btraining_", "", 1)
        if "Btraining_" in fname:
            # If there's a second "Btraining_", remove it without adding an extra underscore
            fname = fname.replace("Btraining_", "")
        else:
            # Add an underscore after the first underscore for single "Btraining_" cases
            fname = fname.replace("_", "__", 1)
    return fname

# Apply the adjustment to the `fname` column
metadata["fname"] = metadata["fname"].apply(adjust_filename)

# Update the `file_path` column to reflect the adjusted `fname`
metadata["file_path"] = metadata["fname"].apply(lambda fname: os.path.join(COMBINED_FOLDER, fname))

# Save the updated metadata back to the CSV
metadata.to_csv(METADATA_FILE, index=False)

print(f"Updated metadata saved to {METADATA_FILE}.")
```

**Data Prepration 3: Augment the Data in the Training Set**

```python
import os
import numpy as np
import pandas as pd
import librosa
import soundfile as sf

# Path to the metadata file
COMBINED_FOLDER = "/root/.cache/kagglehub/datasets/kinguistics/heartbeat-sounds/versions/1/combined_data"
METADATA_FILE = os.path.join(COMBINED_FOLDER, "combined_dataset.csv")

# Load the existing metadata
if os.path.exists(METADATA_FILE):
    metadata = pd.read_csv(METADATA_FILE)
else:
    raise FileNotFoundError(f"Metadata file not found at {METADATA_FILE}.")

# Augment files and update metadata
augmented_files = []
new_metadata_rows = []

# Loop through the metadata and augment files
for index, row in metadata.iterrows():
    fname = row['fname']
    file_path = row['file_path']
    label = row['label']
    sublabel = row['sublabel']

    # Check if the current file is in train_files
    if fname not in train_files:
        continue

    if not os.path.exists(file_path):
        print(f"File not found: {file_path}. Skipping...")
        continue

    # Load the audio file
    try:
        audio, sr = librosa.load(file_path, sr=None)
    except Exception as e:
        print(f"Error loading {file_path}: {e}. Skipping...")
        continue

    # Perform augmentations
    augmentations = {
        "noise": audio + 0.02 * np.random.randn(len(audio)),
        "time_stretch": librosa.effects.time_stretch(audio, rate=1.1),
        "time_stretch_2": librosa.effects.time_stretch(audio, rate=0.8),
        # ensure pitch_shift doesn't completely alter audios
        "pitch_shift": librosa.effects.pitch_shift(audio, sr=sr, n_steps=2),
        "time_shift": np.roll(audio, int(len(audio) * np.random.uniform(-0.2, 0.2))),
        "volume": audio * np.random.uniform(0.8, 1.2)
    }

    # Save augmented files and add to metadata
    for aug_type, augmented_audio in augmentations.items():
        # Generate new file name and path
        new_fname = f"{os.path.splitext(fname)[0]}_{aug_type}.wav"
        new_file_path = os.path.join(COMBINED_FOLDER, new_fname)

        # Save the augmented audio
        sf.write(new_file_path, augmented_audio, sr)

        # Append new file to the augmented files list
        augmented_files.append(new_fname)

        # Create metadata for the augmented file
        new_metadata_rows.append({
            "dataset": row["dataset"],
            "fname": new_fname,
            "label": label,
            "sublabel": sublabel,
            "file_path": new_file_path
        })

# Add new metadata to the DataFrame
if new_metadata_rows:
    new_metadata = pd.DataFrame(new_metadata_rows)
    metadata = pd.concat([metadata, new_metadata], ignore_index=True)

# Save the updated metadata to the CSV file
metadata.to_csv(METADATA_FILE, index=False)

# Output debugging info
print("Augmentation complete.")
print(f"Number of augmented files: {len(augmented_files)}")
print(f"Total files in training set: {len(metadata)}")

# Sanity check: Print sample metadata
print("Sample of updated metadata:")
print(metadata.tail())
```
